### Forecasting menggunakan ARIMA dan Auto ARIMA

```python
# ======================
# 1. Import Library
# ======================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import scipy.stats as stats
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf
from sklearn.metrics import mean_absolute_error, mean_squared_error
from pandas.plotting import register_matplotlib_converters

# Registrasi konverter tanggal agar kompatibel dengan matplotlib
register_matplotlib_converters()
```

# Pre-Processing Data

```python
# ======================
# 2. Unduh & Preprocessing Data USD/IDR
# ======================
ticker = 'USDIDR=X'  # Simbol Yahoo Finance untuk kurs USD ke IDR
start_date = '2020-01-01'
end_date = '2024-04-01'
interval = '1d'

# Mengunduh data menggunakan yfinance
df = yf.download(ticker, start=start_date, end=end_date, interval=interval)

# Hanya gunakan kolom 'Close' (harga penutupan)
df = df[['Close']].dropna()

# Set indeks sebagai tanggal
df.index.name = 'Date'

# Pastikan data bersifat harian (daily frequency)
df = df.asfreq('D')  # pastikan daily

# Mengisi nilai NaN karena hari libur dengan metode forward fill
df = df.fillna(method='ffill')  # isi NaN dari hari libur

print(f"Jumlah data total: {len(df)}")  # debug
```

```python
# ======================
# 3. Split Data (Training vs Testing)
# ======================

# Ambil semua data kecuali 30 hari terakhir sebagai data latih
train = df['Close'][:-30]

# Ambil 30 hari terakhir sebagai data uji
test = df['Close'][-30:]
```

# Soal 2 : ARIMA

```python
# ======================
# 4. Bangun & Latih Model ARIMA
# ======================

# Membuat model ARIMA dengan parameter (p=7, d=2, q=8)
# Catatan: parameter ini bisa didapat dari tunning mandiri
model = ARIMA(train, order=(7, 2, 8))

# Latih model dengan data latih
model_fit = model.fit()
```

```python
# ======================
# 5. Prediksi 30 Hari ke Depan
# ======================

# Lakukan prediksi sebanyak 30 langkah ke depan (30 hari)
forecast = model_fit.forecast(steps=30)

# Pastikan forecast adalah Series & index sama
# Ubah hasil prediksi menjadi Series dan sesuaikan index-nya dengan test
forecast = pd.Series(forecast.values, index=test.index)
```

```python
# ======================
# 6. Visualisasi Full Historis + Testing + Forecast
# ======================
plt.figure(figsize=(14, 6))
plt.plot(train, label=' Data Latih (Train)', color='green')
plt.plot(test, label='Data Aktual (Test)', color='blue')
plt.plot(forecast, label='Prediksi ARIMA (Forecast)', color='red')
plt.title(' USD/IDR - Historis, Validasi, dan Prediksi ARIMA')
plt.xlabel('Tanggal')
plt.ylabel('Kurs USD/IDR')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ======================
# 7. Evaluasi Akurasi Prediksi
# ======================

# Hitung MAE (Mean Absolute Error)
mae = mean_absolute_error(test, forecast)

# Hitung RMSE (Root Mean Squared Error)
rmse = np.sqrt(mean_squared_error(test, forecast))


print(f"MAE  (Mean Absolute Error): {mae:.2f}")
print(f"RMSE (Root Mean Squared Error): {rmse:.2f}")
```

```python
# ======================
# 8. Validasi Residual (Error)
# ======================

# Pastikan test adalah Series
if isinstance(test, pd.DataFrame):
    test = test.squeeze()
    
# Pastikan forecast adalah Series juga
if isinstance(forecast, pd.DataFrame):
    forecast = forecast.squeeze()

# Hitung residual (selisih antara data aktual dan prediksi)
residuals = test - forecast

# Debug 
print("DEBUG: Residual Info")
print(f"Residual type: {type(residuals)}")
print(f"Ada NaN? {residuals.isna().any()}")
print(residuals.head())
```

```python
# Plot Residual Time Series
plt.figure(figsize=(12, 4))
plt.plot(residuals, marker='o', linestyle='-', color='purple')
plt.axhline(0, color='black', linestyle='--')
plt.title('Residual Error (Data Aktual - Prediksi)')
plt.xlabel('Tanggal')
plt.ylabel('Selisih Kurs')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Histogram Residual
plt.figure(figsize=(8, 4))
if residuals.nunique() > 1:
    sns.histplot(residuals, kde=True, color='orange')
    plt.title('Histogram Residual + KDE')
else:
    sns.histplot(residuals, kde=False, color='orange')
    plt.title('Histogram Residual (Tanpa KDE, data tidak variatif)')
plt.xlabel('Residual')
plt.tight_layout()
plt.show()
```

```python
# QQ Plot Residual
plt.figure(figsize=(6, 6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title('QQ Plot Residual (Normality Check)')
plt.tight_layout()
plt.show()
```

```python
# ACF Plot Residual
plt.figure(figsize=(10, 4))
max_lags = min(20, len(residuals)-1)  # otomatis disesuaikan agar tidak error
plot_acf(residuals, lags=max_lags)
plt.title('ACF Plot Residual (Autokorelasi Error)')
plt.tight_layout()
plt.show()
```

```python
# Validasi Plot Aktual-Prediksi
plt.figure(figsize=(12, 5))
plt.plot(test, label="Data Aktual", color='blue')
plt.plot(forecast, label="Prediksi Manual ARIMA (7,2,8)", color='red', linestyle='--', marker='o')
plt.title("Validasi Prediksi Manual ARIMA vs Data Aktual")
plt.xlabel("Tanggal")
plt.ylabel("Kurs USD/IDR")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# SOAL 3 : AUTO ARIMA

```python
from pmdarima import auto_arima

# ======================
# AUTO ARIMA
# ======================
auto_model = auto_arima(
    train,
    start_p=1, start_q=1,
    max_p=10, max_q=10,
    d=2,  # Paksa differencing dua kali, biar adil dengan manual
    seasonal=False,
    stepwise=True,
    suppress_warnings=True,
    error_action='ignore',
    trace=True
)
```

```python
print("Ringkasan Model Auto ARIMA:")
print(auto_model.summary())
```

```python
# Buat model ARIMA dari hasil auto_arima
auto_arima_model = ARIMA(train, order=auto_model.order)
auto_arima_fit = auto_arima_model.fit()

# Prediksi 30 hari ke depan dengan auto_arima
auto_forecast = auto_arima_fit.forecast(steps=30)
auto_forecast = pd.Series(auto_forecast.values, index=test.index)
```

```python
# ======================
# Validasi Residual - AUTO ARIMA
# ======================

# Hitung residual Auto ARIMA
residuals_auto = test - auto_forecast 

# Debug residual
print("DEBUG: Residual Auto ARIMA")
print(f"Residual type: {type(residuals_auto)}")
print(f"Ada NaN? {residuals_auto.isna().any()}")
print(residuals_auto.head())

# Plot Residual Time Series
plt.figure(figsize=(12, 4))
plt.plot(residuals_auto, marker='o', linestyle='-', color='darkcyan')
plt.axhline(0, color='black', linestyle='--')
plt.title('Residual Auto ARIMA (Test - Forecast)')
plt.xlabel('Tanggal')
plt.ylabel('Error')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Histogram Residual Auto ARIMA
plt.figure(figsize=(8, 4))
if residuals_auto.nunique() > 1:
    sns.histplot(residuals_auto, kde=True, color='teal')
    plt.title('Histogram Residual Auto ARIMA + KDE')
else:
    sns.histplot(residuals_auto, kde=False, color='teal')
    plt.title('Histogram Residual Auto ARIMA (Tanpa KDE)')
plt.xlabel('Residual')
plt.tight_layout()
plt.show()
```

```python
# QQ Plot Residual Auto ARIMA
plt.figure(figsize=(6, 6))
stats.probplot(residuals_auto, dist="norm", plot=plt)
plt.title('QQ Plot Residual Auto ARIMA')
plt.tight_layout()
plt.show()
```

```python
# ACF Plot Residual Auto ARIMA
plt.figure(figsize=(10, 4))
max_lags = min(20, len(residuals_auto)-1)
plot_acf(residuals_auto, lags=max_lags)
plt.title('ACF Plot Residual Auto ARIMA')
plt.tight_layout()
plt.show()
```

```python
# Validasi Plot Aktual-Prediksi Auto ARIMA
plt.figure(figsize=(12, 5))
plt.plot(test, label="Data Aktual", color='blue')
plt.plot(auto_forecast, label=f"Prediksi Auto ARIMA {auto_model.order}", color='orange', linestyle='--', marker='o')
plt.title("Validasi Prediksi Auto ARIMA vs Data Aktual")
plt.xlabel("Tanggal")
plt.ylabel("Kurs USD/IDR")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# PERBANDINGAN EVALUASI : Manual vs Auto ARIMA

```python
# Manual
mae_manual = mean_absolute_error(test, forecast)
rmse_manual = np.sqrt(mean_squared_error(test, forecast))

#Auto
mae_auto = mean_absolute_error(test, auto_forecast)
rmse_auto = np.sqrt(mean_squared_error(test, auto_forecast))

print("PERBANDINGAN AKURASI:")
print(f"Manual ARIMA (7,2,8): MAE = {mae_manual:.2f}, RMSE = {rmse_manual:.2f}")
print(f"Auto ARIMA {auto_model.order}: MAE = {mae_auto:.2f}, RMSE = {rmse_auto:.2f}")
```

```python
# ======================
# Visualisasi Perbandingan Manual vs Auto ARIMA
# ======================
plt.figure(figsize=(14, 6))
plt.plot(test, label='Data Aktual (Test)', color='blue')
plt.plot(forecast, label='Manual ARIMA (7,2,8)', color='red')
plt.plot(auto_forecast, label=f'Auto ARIMA {auto_model.order}', color='orange', linestyle='--')
plt.title('Perbandingan Prediksi Manual vs Auto ARIMA')
plt.xlabel('Tanggal')
plt.ylabel('Kurs USD/IDR')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
