```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller

# Fungsi utama untuk menjalankan seluruh proses
def main(file_path):
    # Langkah 1: Memuat data CSV
    data = pd.read_csv(file_path, header=2)
    
    # Mengatur nama kolom yang benar
    data.columns = ['Date', 'Close', 'High', 'Low', 'Open', 'Volume']
    
    # Menghapus baris yang tidak relevan (metadata)
    data = data.dropna(subset=['Date'])  # Menghapus baris dengan 'Date' kosong
    data = data[data['Date'] != 'Date']  # Menghapus baris dengan nilai 'Date' sebagai string

    # Mengonversi kolom 'Date' ke format datetime
    data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

    # Menghapus baris dengan nilai NaT pada kolom 'Date' setelah konversi
    data.dropna(subset=['Date'], inplace=True)

    # Menetapkan kolom 'Date' sebagai indeks
    data.set_index('Date', inplace=True)
    
    # Membuat DatetimeIndex dengan frekuensi harian yang terdefinisi
    data = data.asfreq('D')  # Menetapkan frekuensi harian secara eksplisit

    # Mengisi missing values dengan forward fill (ffill)
    data.ffill(inplace=True)

    # Memeriksa dan menangani duplikasi
    data.drop_duplicates(inplace=True)
    
    # Langkah 2: Memeriksa stasionaritas dan melakukan differencing jika diperlukan
    adf_result = adfuller(data['Close'])
    if adf_result[1] > 0.05:
        print("Data tidak stasioner, melakukan differencing.")
        data_diff = data['Close'].diff().dropna()  # Differencing pertama untuk stasioneritas
    else:
        print("Data sudah stasioner.")
        data_diff = data['Close']

    # Langkah 3: Membagi data menjadi training dan testing set
    train_size = int(len(data_diff) * 0.8)
    train, test = data_diff[:train_size], data_diff[train_size:]

    # Langkah 4: Membangun dan melatih model ARIMA
    model = ARIMA(train, order=(1, 0, 1))  # ARIMA(p=1, d=0, q=1)
    model_fit = model.fit()

    # Langkah 5: Membuat peramalan untuk testing set
    forecast = model_fit.forecast(steps=len(test))

    # Memastikan panjang forecast sesuai dengan panjang data test
    forecast_index = data.index[len(train):len(train)+len(test)]

    # Visualisasi perbandingan antara data aktual dan peramalan
    plt.figure(figsize=(12, 6))
    
    # Plot data asli dan peramalan
    plt.plot(data.index[:len(train)], train, label='Training Data')
    plt.plot(forecast_index, test, label='Test Data', color='orange')
    plt.plot(forecast_index, forecast, label='Forecast', color='red', linestyle='--')
    
    plt.title('Forecasting Stock Prices using ARIMA')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.show()

# Path ke file CSV yang akan diproses
file_path = 'ihsg_data.csv'

# Menjalankan proses utama
main(file_path)
```
