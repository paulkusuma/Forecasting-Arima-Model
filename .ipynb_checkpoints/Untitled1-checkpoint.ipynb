```python
# ======================
#  1. Import Library
# ======================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
from pandas.plotting import register_matplotlib_converters

register_matplotlib_converters()

# ======================
#  2. Load & Preprocessing CSV IHSG
# ======================

# Baca file CSV (file memiliki 2 baris awal non-data)
df = pd.read_csv('ihsg_data.csv')

# Hapus dua baris pertama yang merupakan header tambahan
df = df.iloc[2:].copy()

# Ganti nama kolom 'Price' menjadi 'Date' (karena berisi tanggal)
df.rename(columns={'Price': 'Date'}, inplace=True)

# Ambil hanya kolom yang dibutuhkan
df = df[['Date', 'Close']]

# Konversi tipe data
df['Date'] = pd.to_datetime(df['Date'])     # Kolom tanggal menjadi datetime
df['Close'] = pd.to_numeric(df['Close'])    # Kolom harga ditransform ke float

# Jadikan kolom Date sebagai index time series
df.set_index('Date', inplace=True)

# Sortir data berdasarkan tanggal (penting untuk time series)
df = df.sort_index()

# ======================
#  3. Split Train dan Test
# ======================

# Gunakan 30 hari terakhir sebagai data test (validasi)
train = df['Close'][:-30]  # Semua data kecuali 30 hari terakhir
test = df['Close'][-30:]   # Data 30 hari terakhir

# ======================
#  4. Bangun & Latih Model ARIMA
# ======================

# Gunakan parameter ARIMA (p=7, d=1, q=2)
model = ARIMA(train, order=(7, 1, 2))
model_fit = model.fit()

# Tampilkan ringkasan model
print(model_fit.summary())

# ======================
#  5. Prediksi Data Test (Validasi)
# ======================

# Prediksi sebanyak panjang data test (30 hari)
forecast = model_fit.forecast(steps=30)

# ======================
#  6. Visualisasi: Prediksi vs Data Asli
# ======================

plt.figure(figsize=(12, 5))
plt.plot(test.index, test.values, label='Data Asli (Test)', color='blue')
plt.plot(test.index, forecast.values, label='Prediksi ARIMA', color='red')
plt.title(' Prediksi vs Data Asli IHSG (30 Hari Validasi)')
plt.xlabel('Tanggal')
plt.ylabel('Harga IHSG')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ======================
#  7. Evaluasi Error Prediksi
# ======================
# Hitung MAE & RMSE
mae = mean_absolute_error(test, forecast)  # Mean Absolute Error
rmse = np.sqrt(mean_squared_error(test, forecast))  # Root Mean Squared Error

# Tampilkan hasil evaluasi
print(f"\n MAE  (Mean Absolute Error): {mae:.2f}")
print(f" RMSE (Root Mean Squared Error): {rmse:.2f}")

# ======================
#  8. Visualisasi Residual (Error Model)
# ======================

# Residual = selisih antara hasil prediksi dengan data aktual (selama training)
residuals = model_fit.resid

plt.figure(figsize=(12, 4))
plt.plot(residuals)
plt.title(' Visualisasi Residual ARIMA (Training Set)')
plt.xlabel('Tanggal')
plt.ylabel('Error / Residual')
plt.grid(True)
plt.tight_layout()
plt.show()

# ======================
#  9. Prediksi 30 Hari ke Depan dari Model Final
# ======================

# Bangun ulang model dengan seluruh data
final_model = ARIMA(df['Close'], order=(7, 1, 2))
final_fit = final_model.fit()

# Prediksi 30 hari ke depan (forecast masa depan)
future_forecast = final_fit.forecast(steps=30)

# Buat index tanggal untuk forecast ke depan
last_date = df.index[-1]
future_index = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=30, freq='D')
future_series = pd.Series(future_forecast.values, index=future_index)

# Visualisasi prediksi ke depan
plt.figure(figsize=(12, 5))
plt.plot(df['Close'], label='Data Historis')
plt.plot(future_series.index, future_series.values, color='red', label='Prediksi 30 Hari ke Depan')
plt.title(' Prediksi IHSG (30 Hari ke Depan)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Cetak nilai prediksi
print("\n Prediksi IHSG 30 Hari Mendatang:")
print(future_series)
```

```python
# ======================
#  1. Import Library
# ======================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import scipy.stats as stats
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf
from sklearn.metrics import mean_absolute_error, mean_squared_error
from pandas.plotting import register_matplotlib_converters

register_matplotlib_converters()

# ======================
#  2. Unduh & Preprocessing Data USD/IDR
# ======================
ticker = 'USDIDR=X'
start_date = '2020-01-01'
end_date = '2024-04-01'
interval = '1d'

df = yf.download(ticker, start=start_date, end=end_date, interval=interval)
df = df[['Close']].dropna()
df.index.name = 'Date'
df = df.asfreq('D')  # pastikan daily
df = df.fillna(method='ffill')  # isi NaN dari hari libur

print(f"Jumlah data total: {len(df)}")  # debug

# ======================
#  3. Split Data (Training vs Testing)
# ======================
train = df['Close'][:-30]
test = df['Close'][-30:]

# ======================
#  4. Bangun & Latih Model ARIMA
# ======================
model = ARIMA(train, order=(7, 2, 8))
model_fit = model.fit()

# ======================
#  5. Prediksi 30 Hari ke Depan
# ======================
forecast = model_fit.forecast(steps=30)

#  Pastikan forecast adalah Series & index sama
forecast = pd.Series(forecast.values, index=test.index)

# ======================
#  6. Visualisasi Full Historis + Testing + Forecast
# ======================
plt.figure(figsize=(14, 6))
plt.plot(train, label=' Data Latih (Train)', color='green')
plt.plot(test, label=' Data Aktual (Test)', color='blue')
plt.plot(forecast, label=' Prediksi ARIMA (Forecast)', color='red')
plt.title(' USD/IDR - Historis, Validasi, dan Prediksi ARIMA')
plt.xlabel('Tanggal')
plt.ylabel('Kurs USD/IDR')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ======================
#  7. Evaluasi Akurasi Prediksi
# ======================
mae = mean_absolute_error(test, forecast)
rmse = np.sqrt(mean_squared_error(test, forecast))
print(f"\n MAE  (Mean Absolute Error): {mae:.2f}")
print(f" RMSE (Root Mean Squared Error): {rmse:.2f}")

# ======================
#  8. Validasi Residual (Error)
# ======================

# Pastikan test adalah Series
if isinstance(test, pd.DataFrame):
    test = test.squeeze()

# Pastikan forecast adalah Series juga
if isinstance(forecast, pd.DataFrame):
    forecast = forecast.squeeze()
    
residuals = test - forecast

#  Debug check
print("\n DEBUG: Residual Info")
print(f"Residual type: {type(residuals)}")
print(f"Ada NaN? {residuals.isna().any()}")
print(residuals.head())

#  Plot Residual Time Series
plt.figure(figsize=(12, 4))
plt.plot(residuals, marker='o', linestyle='-', color='purple')
plt.axhline(0, color='black', linestyle='--')
plt.title(' Residual Error (Data Aktual - Prediksi)')
plt.xlabel('Tanggal')
plt.ylabel('Selisih Kurs')
plt.grid(True)
plt.tight_layout()
plt.show()

#  Histogram Residual
plt.figure(figsize=(8, 4))
if residuals.nunique() > 1:
    sns.histplot(residuals, kde=True, color='orange')
    plt.title(' Histogram Residual + KDE')
else:
    sns.histplot(residuals, kde=False, color='orange')
    plt.title(' Histogram Residual (Tanpa KDE, data tidak variatif)')
plt.xlabel('Residual')
plt.tight_layout()
plt.show()

#  QQ Plot Residual
plt.figure(figsize=(6, 6))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title(' QQ Plot Residual (Normality Check)')
plt.tight_layout()
plt.show()

#  ACF Plot Residual
plt.figure(figsize=(10, 4))
max_lags = min(20, len(residuals)-1)  # otomatis disesuaikan agar tidak error
plot_acf(residuals, lags=max_lags)
plt.title(' ACF Plot Residual (Autokorelasi Error)')
plt.tight_layout()
plt.show()

plt.plot(test, label="Aktual")
plt.plot(forecast, label="Prediksi")
plt.legend()
plt.show()


# ======================
#  Tambahan: AUTO ARIMA
# ======================
from pmdarima import auto_arima

auto_model = auto_arima(
    train,
    start_p=1, start_q=1,
    max_p=10, max_q=10,
    d=2,  # Paksa differencing dua kali
    seasonal=False,
    stepwise=True,
    suppress_warnings=True,
    error_action='ignore',
    trace=True
)

print("\n Ringkasan Model Auto ARIMA:")
print(auto_model.summary())

#  Buat model ARIMA dari hasil auto_arima
auto_arima_model = ARIMA(train, order=auto_model.order)
auto_arima_fit = auto_arima_model.fit()

#  Prediksi 30 hari ke depan dengan auto_arima
auto_forecast = auto_arima_fit.forecast(steps=30)
auto_forecast = pd.Series(auto_forecast.values, index=test.index)


# ======================
#  Validasi Residual - AUTO ARIMA
# ======================

# Hitung residual Auto ARIMA
residuals_auto = test - auto_forecast 

#  Debug residual
print("\n DEBUG: Residual Auto ARIMA")
print(f"Residual type: {type(residuals_auto)}")
print(f"Ada NaN? {residuals_auto.isna().any()}")
print(residuals_auto.head())

#  Plot Residual Time Series
plt.figure(figsize=(12, 4))
plt.plot(residuals_auto, marker='o', linestyle='-', color='darkcyan')
plt.axhline(0, color='black', linestyle='--')
plt.title(' Residual Auto ARIMA (Test - Forecast)')
plt.xlabel('Tanggal')
plt.ylabel('Error')
plt.grid(True)
plt.tight_layout()
plt.show()

#  Histogram Residual Auto ARIMA
plt.figure(figsize=(8, 4))
if residuals_auto.nunique() > 1:
    sns.histplot(residuals_auto, kde=True, color='teal')
    plt.title(' Histogram Residual Auto ARIMA + KDE')
else:
    sns.histplot(residuals_auto, kde=False, color='teal')
    plt.title(' Histogram Residual Auto ARIMA (Tanpa KDE)')
plt.xlabel('Residual')
plt.tight_layout()
plt.show()

#  QQ Plot Residual Auto ARIMA
plt.figure(figsize=(6, 6))
stats.probplot(residuals_auto, dist="norm", plot=plt)
plt.title(' QQ Plot Residual Auto ARIMA')
plt.tight_layout()
plt.show()

#  ACF Plot Residual Auto ARIMA
plt.figure(figsize=(10, 4))
max_lags = min(20, len(residuals_auto)-1)
plot_acf(residuals_auto, lags=max_lags)
plt.title(' ACF Plot Residual Auto ARIMA')
plt.tight_layout()
plt.show()



# ======================
#  Bandingkan Evaluasi: Manual vs Auto ARIMA
# ======================
mae_manual = mean_absolute_error(test, forecast)
rmse_manual = np.sqrt(mean_squared_error(test, forecast))

mae_auto = mean_absolute_error(test, auto_forecast)
rmse_auto = np.sqrt(mean_squared_error(test, auto_forecast))

print("\n PERBANDINGAN AKURASI:")
print(f" Manual ARIMA (7,2,8): MAE = {mae_manual:.2f}, RMSE = {rmse_manual:.2f}")
print(f" Auto ARIMA {auto_model.order}: MAE = {mae_auto:.2f}, RMSE = {rmse_auto:.2f}")

# ======================
#  Visualisasi Perbandingan Manual vs Auto ARIMA
# ======================
plt.figure(figsize=(14, 6))
plt.plot(test, label=' Data Aktual (Test)', color='blue')
plt.plot(forecast, label=' Manual ARIMA (7,2,8)', color='red')
plt.plot(auto_forecast, label=f' Auto ARIMA {auto_model.order}', color='orange', linestyle='--')
plt.title(' Perbandingan Prediksi Manual vs Auto ARIMA')
plt.xlabel('Tanggal')
plt.ylabel('Kurs USD/IDR')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
