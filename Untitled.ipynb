```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
```

```python
# Memuat data
data = pd.read_csv("ihsg_data.csv")
data
```

```python
# Step 1: Menghapus baris yang tidak relevan
data_cleaned = data.iloc[2:].reset_index(drop=True)
data_cleaned
```

```python
# Step 2: Mengonversi kolom 'Price' menjadi kolom 'Date'
data_cleaned['Date'] = pd.to_datetime(data_cleaned['Price'], errors='coerce')
```

```python
data_cleaned
```

```python
# Step 3: Mengonversi kolom lainnya ke numerik dan menangani kesalahan dengan 'coerce'
price_columns = ['Close', 'High', 'Low', 'Open', 'Volume']
for col in price_columns:
    data_cleaned[col] = pd.to_numeric(data_cleaned[col], errors='coerce')
```

```python
# Step 3: Menghapus kolom 'Price' karena sudah tidak relevan
data_cleaned = data_cleaned.drop(columns=['Price'])
data_cleaned
```

```python
# Step 4: Membatasi jumlah digit desimal pada kolom yang relevan (misalnya 2 digit desimal)
decimal_places = 2  # Anda bisa mengubah jumlah desimal sesuai kebutuhan
price_columns = ['Close', 'High', 'Low', 'Open', 'Volume']
```

```python
for col in price_columns:
    data_cleaned[col] = data_cleaned[col].round(decimal_places)
```

```python
# Step 5: Menangani missing values (jika ada)
data_cleaned = data_cleaned.interpolate(method='linear', axis=0)
```

```python
# Step 6: Menjadikan kolom 'Date' sebagai index
data_cleaned.set_index('Date', inplace=True)
```

```python
# Menampilkan data setelah diproses
print(data_cleaned.head())
```

```python
# Step 1: Memeriksa stasioneritas data menggunakan uji Augmented Dickey-Fuller (ADF)
result = adfuller(data_cleaned['Close'].dropna())
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")
if result[1] > 0.05:
    print("Data tidak stasioner, perlu dilakukan differencing.")
else:
    print("Data sudah stasioner.")
```

```python
# Step 2: Melakukan differencing jika data tidak stasioner
# Jika data tidak stasioner, kita akan lakukan differencing untuk menghilangkan tren
data_cleaned['Close_diff'] = data_cleaned['Close'].diff().dropna()
```

```python
# Step 3: Menentukan parameter p, d, q menggunakan ACF dan PACF plot
plot_acf(data_cleaned['Close_diff'].dropna(), lags=40)
plot_pacf(data_cleaned['Close_diff'].dropna(), lags=40)
plt.show()
```

```python
# Berdasarkan plot ACF dan PACF, tentukan p, d, q.
# Misalnya p=1, d=1, q=1 untuk model ARIMA(1,1,1)
p, d, q = 5, 1, 2
```

```python
# Step 4: Membangun model ARIMA
model = ARIMA(data_cleaned['Close'], order=(p, d, q))
model_fit = model.fit()
```

```python
# Step 5: Mengevaluasi model
print(model_fit.summary())
```

```python
# Step 6: Forecasting
forecast_steps = 90  # Tentukan jumlah langkah prediksi
forecast = model_fit.forecast(steps=forecast_steps)
```

```python
# Menampilkan hasil forecasting
forecast_dates = pd.date_range(start=data_cleaned.index[-1], periods=forecast_steps+1, freq='B')[1:]

forecast_series = pd.Series(forecast, index=forecast_dates)
```

```python
# Plot hasil forecast
plt.figure(figsize=(10,6))
plt.plot(data_cleaned['Close'], label='Data Aktual')
plt.plot(forecast_series, label='Forecast', color='red')
plt.title('ARIMA Forecasting')
plt.legend()
plt.show()
```
